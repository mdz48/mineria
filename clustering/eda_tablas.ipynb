```python
import pandas as pd
import numpy as np
import os
from IPython.display import display

pd.set_option('display.max_columns', None)

base_path = 'data' if os.path.exists('data') else 'clustering/data'
df_sintetico = pd.read_csv(os.path.join(base_path, 'datos_embarazo_sintetico.csv'))
df_meta = pd.read_csv(os.path.join(base_path, 'metadatos_ground_truth.csv'))
df = df_sintetico.merge(df_meta, on='paciente_id')
display(df.head())
```

```python
info_df = pd.DataFrame({
    'Tipo': df_sintetico.dtypes,
    'No Nulos': df_sintetico.notna().sum(),
    'Nulos': df_sintetico.isna().sum(),
    '% Nulos': (df_sintetico.isna().sum() / len(df_sintetico) * 100).round(2)
})
display(info_df)
```

```python
display(df_sintetico.describe().round(2))
```

```python
display(df_sintetico.describe(include=['object']).T)
```

```python
numeric_cols = df_sintetico.select_dtypes(include=[np.number]).columns.drop('paciente_id', errors='ignore')
cluster_profile = df.groupby('cluster_verdadero')[numeric_cols].mean().round(2)
display(cluster_profile)
```

```python
outlier_counts = df_meta['tipo_outlier'].value_counts().reset_index()
outlier_counts.columns = ['Tipo de Outlier', 'Frecuencia']
outlier_counts['Porcentaje'] = (outlier_counts['Frecuencia'] / len(df_meta) * 100).round(2)
display(outlier_counts)
```
